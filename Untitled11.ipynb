```python
spam_messages = messages[messages["v1"] == "spam"]["v2"]
ham_messages = messages[messages["v1"] == "ham"]["v2"]

spam_words = []
ham_words = []

# Iterate through spam messages and tokenize
for message in spam_messages:
    # Tokenize the message and extend the spam_words list
    tokens = word_tokenize(message)
    spam_words.extend(tokens)

# (Optional) Similarly, process ham_messages if needed
for message in ham_messages:
    tokens = word_tokenize(message)
    ham_words.extend(tokens)
```

```python
messages["messageLength"] = messages["v2"].apply(len) # Changed 'text' to 'v2'
messages["messageLength"].describe()
```

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("english")

def cleanText(message):

    message = message.translate(str.maketrans('', '', string.punctuation))
    words = [stemmer.stem(word) for word in message.split() if word.lower() not in stopwords.words("english")]

    return " ".join(words)

messages["text"] = messages["text"].apply(cleanText)
messages.head(n = 10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer(encoding = "latin-1", strip_accents = "unicode", stop_words = "english")
features = vec.fit_transform(messages["text"])
print(features.shape)
```

```python
def encodeCategory(cat):
    if cat == "spam":
        return 1
    else:
        return 0

# Assuming the column containing spam/ham labels is named 'v1'
messages["category"] = messages["v1"].apply(encodeCategory)  # Changed 'category' to 'v1'
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, messages["category"], stratify = messages["category"], test_size = 0.2)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import fbeta_score

from sklearn.naive_bayes import MultinomialNB
gaussianNb = MultinomialNB()
gaussianNb.fit(X_train, y_train)

y_pred = gaussianNb.predict(X_test)

print(fbeta_score(y_test, y_pred, beta = 0.5))
```
